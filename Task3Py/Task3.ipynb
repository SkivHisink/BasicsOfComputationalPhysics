```python
import numpy as numpy
import matplotlib.pyplot as plt
import math
from scipy.misc import derivative

%matplotlib inline
```

```python
def Trapezoid_method(function, minimum= float('-inf'), maximum = float('inf'), partsNum = 2) -> float:
    step = (maximum-minimum)/partsNum
    result = 0.0
    for i in range(partsNum):
        begin = minimum + step * i
        end = minimum + step * (i + 1)
        result += (function(begin)+ function(end)) * 0.5 * (end - begin)
    return result
def Simpson_method(function, minimum= float('-inf'), maximum = float('inf'), partsNum = 2) -> float:
    step = (maximum-minimum)/partsNum
    result = 0.0
    for i in range(0, partsNum-1, 2):
        left = minimum + step * i
        middle = minimum + step * (i + 1)
        right = minimum + step * (i + 2)
        result += (right - left) / 6.0 * (function(left) + 4.0 * function(middle) + function(right))
    return result
```

```python
def function_a(x):
    return 1 / (1 + math.pow(x, 2))
trapezoid = list()
simpson = list()
for i in range(4, 257, 4):
    print('Interval devided on ' + str(i)+' parts')
    trapezoid_r = Trapezoid_method(function_a, -1, 1, i)
    trapezoid.append(trapezoid_r)
    print('Trapezoid - ' + str(trapezoid_r))
    simpson_r = Simpson_method(function_a, -1, 1, i)
    simpson.append(simpson_r)
    print('Simpson - ' + str(simpson_r))
print('Write answer is ' + str(numpy.pi /2))
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
fig, ax = plt.subplots()
ax.plot(trapezoid, label='Trapezoid')
ax.plot(simpson, label='Simpson')
```

```python
def function_b(x):
    return x ** (1/3) * numpy.exp(math.sin(x))
trapezoid = list()
simpson = list()
for i in range(4, 257, 4):
    print('Interval devided on ' + str(i)+' parts')
    trapezoid_r = Trapezoid_method(function_b, 0, 1, i)
    trapezoid.append(trapezoid_r)
    print('Trapezoid - ' + str(trapezoid_r))
    simpson_r = Simpson_method(function_b, 0, 1, i)
    simpson.append(simpson_r)
    print('Simpson - ' + str(simpson_r))
print('Write answer is ' + '1.29587400873170808779')
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
fig, ax = plt.subplots()
ax.plot(trapezoid, label='Trapezoid')
ax.plot(simpson, label='Simpson')
```
