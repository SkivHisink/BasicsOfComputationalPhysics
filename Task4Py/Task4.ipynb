```python
import numpy as numpy
import matplotlib.pyplot as plt
import math
from scipy.misc import derivative

%matplotlib inline
```

```python
def Trapezoid_method(function, minimum= float('-inf'), maximum = float('inf'), partsNum = 2) -> float:
    step = (maximum-minimum)/partsNum
    result = 0.0
    for i in range(partsNum):
        begin = minimum + step * i
        end = minimum + step * (i + 1)
        result += (function(begin)+ function(end)) * 0.5 * (end - begin)
    return result
def Simpson_method(function, minimum= float('-inf'), maximum = float('inf'), partsNum = 2) -> float:
    step = (maximum-minimum)/partsNum
    result = 0.0
    for i in range(0, partsNum-1, 2):
        left = minimum + step * i
        middle = minimum + step * (i + 1)
        right = minimum + step * (i + 2)
        result += (right - left) / 6.0 * (function(left) + 4.0 * function(middle) + function(right))
    return result
def derivative_(function, x, step, order):
    result = 0.0
    for i in range(1, order+1):
        result+= (function(x + i * step)-function(x - i * step)) / (i * 2.0 * step)
    return result / order
```

```python
def changebleBessel(x, def_step, int_step):
    zero_bessel = BesselFunction(0, Simpson_method, 0, 2 * math.pi, int_step)
    der_zb = derivative_(zero_bessel, x, def_step, 2)
    bessel_val = calculateBessel(1, x, Simpson_method, 0, math.pi, int_step)
    return bessel_val + der_zb
changebleBessel(0, derivativeStep, integralSteps)
```
